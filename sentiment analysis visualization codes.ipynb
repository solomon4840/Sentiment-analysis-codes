```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File path to the data
file_path = r'C:/Users/SOLOMON/Desktop/Python/Flipkart analyzed data.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Summing the sentiment counts
pos_count = df['pos'].sum().astype(int)
neg_count = df['neg'].sum().astype(int)
neutral_count = df['neu'].sum().astype(int)

# Data for visualization
labels = ['Positive', 'Negative', 'Neutral']
counts = [pos_count, neg_count, neutral_count]

# Convert data to DataFrame to use hue with Seaborn
sentiment_data = pd.DataFrame({'Sentiment': labels, 'Counts': counts})

# Creating the bar chart with hue and setting legend=False
sns.barplot(x='Sentiment', y='Counts', data=sentiment_data, hue='Sentiment', 
            palette=['green', 'red', 'gray'], legend=False)

plt.title('Sentiment Distribution Across All Comments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Comments')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path to the data
file_path = r'C:/Users/SOLOMON/Desktop/Python/Flipkart analyzed data.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Summing the sentiment counts
pos_count = df['pos'].sum().astype(int)
neg_count = df['neg'].sum().astype(int)
neutral_count = df['neu'].sum().astype(int)

# Data for the pie chart
labels = ['Positive', 'Negative', 'Neutral']
sizes = [pos_count, neg_count, neutral_count]
colors = ['green', 'red', 'gray']

# Pie chart for sentiment distribution
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Sentiment Distribution Across All Comments')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd

# File path to the data
file_path = r'C:/Users/SOLOMON/Desktop/Python/Flipkart analyzed data.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Summing the sentiment counts
pos_count = df['pos'].sum().astype(int)
neg_count = df['neg'].sum().astype(int)
neutral_count = df['neu'].sum().astype(int)

# Total number of comments
total_comments = 499  # You mentioned the total number of comments is 499

# Calculate the percentage distribution
sentiment_counts = {
    'positive': pos_count,
    'negative': neg_count,
    'neutral': neutral_count
}

sentiment_percentages = {k: (v / total_comments) * 100 for k, v in sentiment_counts.items()}

# Print the percentage distribution
print("Percentage Distribution:")
for sentiment, percentage in sentiment_percentages.items():
    print(f"{sentiment.capitalize()}: {percentage:.2f}%")
```
